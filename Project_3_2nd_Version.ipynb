```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

#remove missing values
missing_values = [".", ",", "na"]
data = pd.read_csv('https://raw.githubusercontent.com/mdgalvez/Phyton_PROJECT-1/master/New_York_City_Leading_Causes_of_Death%20(2).csv', na_values = missing_values)

#filter by specific year
#df_ethnicity =  pd.DataFrame(data,columns=['Leading Cause','Race Ethnicity','Deaths','Year'])
#df_ethnicity_year =  df_ethnicity[df_ethnicity.Year.eq(2010)].groupby(['Leading Cause','Race Ethnicity']).Deaths.sum().reset_index()

#sum all year
df_ethnicity =  pd.DataFrame(data,columns=['Leading Cause','Race Ethnicity','Deaths'])
df_ethnicity_year =  df_ethnicity.groupby(['Leading Cause','Race Ethnicity']).Deaths.sum().reset_index()
df_ethnicity_year["mean_rank"] = df_ethnicity_year.groupby('Leading Cause').Deaths.transform(np.sum).rank(ascending=0,method='dense')
df_ethnicity_year = df_ethnicity_year[df_ethnicity_year['mean_rank'] < 11]
##.sort_values("Deaths", ascending=False).head(10)
#df_ethnicity_year
#df

#df_year = df[df.Year.eq(2010)].sort_values('Death Rate',ascending=False).head(10)
#df_year
chart = sns.catplot(x='Leading Cause', y='Deaths', hue='Race Ethnicity', data=df_ethnicity_year, kind="bar", size=4, aspect=4)

chart.set_xticklabels(rotation=90)

df_sex =  pd.DataFrame(data,columns=['Leading Cause','Sex','Deaths'])
df_sex_allyear =  df_sex .groupby(['Leading Cause','Sex']).Deaths.sum().reset_index()
df_sex_allyear["mean_rank"] = df_sex_allyear.groupby('Leading Cause').Deaths.transform(np.sum).rank(ascending=0,method='dense')
df_sex_allyear = df_sex_allyear[df_sex_allyear['mean_rank'] < 11]

chartSex = sns.catplot(x='Leading Cause', y='Deaths', hue='Sex', data=df_sex_allyear, kind="point", size=4, aspect=4)

chartSex.set_xticklabels(rotation=90)

#df.head(1000)
#df.drop_duplicates(keep = 'first',inplace = True)
#df

#df.max()
#df.min()
#df.count()
#df.describe()
#df['Death Rate'].mean()
#df.hist()
#df.plot()
#df.plot(kind="hist")
#df.plot(kind="bar", x='Year', y='Deaths')
#df.plot(kind='bar')
#plt.xticks(rotation=90)

#df['Age Adjusted Death Rate'].mean()
#df = df.sort_values('Deaths',ascending=False)
#sns.barplot(x = "Year", y ="Deaths", data=df)
#sns.barplot(x='Deaths', y='Year', hue='Leading Cause', data=df, kind='bar')
#plt.xticks(rotation=60)

#sns.barplot(x = "Deaths", y ="Leading Cause", data=df)
#sns.barplot(x = "Sex", y ="Deaths", data=df)
#sns.barplot(x= "Year", y="Deaths", data=df)
#sns.barplot(x = "Race Ethnicity", y ="Deaths", data=df)

#sns.catplot(x="Race Ethnicity", y="Deaths", hue="Leading Cause", kind="bar", data=df);
#plt.xticks(rotation=60)
```
